```python
import sklearn
from sklearn import datasets
```

```python
from sklearn.datasets import load_digits
data = load_digits()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.DataFrame(data.data,columns = data.feature_names)
dataset.head()
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters = 10,random_state = 1)
result = kmeans.fit_predict(dataset)
```

```python
data.target
```

```python
result = pd.DataFrame(result)
original = pd.DataFrame(data.target)
```

```python
print(result.value_counts())
print(original.value_counts())
```

```python
print("mean comparision for original and results: ")
print("original: ",original.mean()[0])
print("result: ", result.mean()[0])
print("     ")
print("     ")
print("std comparision for original and results: ")
print("original: ",original.std()[0])
print("result: ",result.std()[0])
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
vif_data = pd.DataFrame()
vif_data["features"] = dataset.columns
vif_data["VIF value"] = [variance_inflation_factor(dataset.values,i) for i in range(len(dataset.columns))]
```

```python
features_to_be_removed = []
```

```python
for i in range(len(vif_data)):
    if vif_data["VIF value"][i] > 10:
        print(vif_data.iloc[i])
        features_to_be_removed.append(vif_data["features"][i])
```

```python
features_to_be_removed
```

```python
dataset_2 = dataset.drop(features_to_be_removed,axis = 1)
```

```python
kmeans_2 = KMeans(n_clusters = 10,random_state = 1)
result_2 = kmeans_2.fit_predict(dataset_2)
```

```python
result_2 = pd.DataFrame(result_2)
original = pd.DataFrame(data.target)
```

```python
print("mean comparision for original and results: ")
print("original: ",original.mean()[0])
print("result: ", result_2.mean()[0])
print("     ")
print("     ")
print("std comparision for original and results: ")
print("original: ",original.std()[0])
print("result: ",result_2.std()[0])
```
